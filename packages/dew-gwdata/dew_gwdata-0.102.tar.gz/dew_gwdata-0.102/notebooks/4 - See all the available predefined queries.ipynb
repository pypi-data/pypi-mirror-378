## Finding wells

You can search for wells that belong to a certain type of drillhole group. The easiest example is collecting all wells in monitoring networks. Monitoring networks are a drillhole group of type ``OMN``.

```python
wells = db.wells_in_group_type("OMN")
wells.info()
```

```python
wells[["well_id"] + list(wells.columns)[11:]].head()
```

Or you can select wells that belong to a particular group.

First, get the list of current monitoring network group codes from the DataFrame we just loaded...

```python
wells.group_code.unique()
```

I'll pick two from the Upper SE: PADTH and TINT_COON.

```python
wells = db.wells_in_groups(["PADTH", "TINT_COON"])
wells.info()
```

```python
wells[["well_id"] + list(wells.columns[11:])].head()
```
