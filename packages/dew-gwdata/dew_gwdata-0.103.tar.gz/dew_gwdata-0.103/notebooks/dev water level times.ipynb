```python
import dew_gwdata as gd
```

```python
db = gd.sageodata()
```

```python
db.query(
    "select table_name, column_name, data_type from all_tab_columns where table_name = 'WA_WATER_LEVEL_VW'"
)
```

```python
dh_no = db.find_wells("NAR070").dh_no[0]
sql = f"""
select 
obs_date, depth_to_water, created_by, creation_date, modified_by, modified_date 
from dhdb.wa_water_level_vw 
where drillhole_no = {dh_no}
"""
cursor = db.cursor()
cursor.execute(sql)
for row in cursor:
    if row[1]:
        print("obs_date {}".format(row[0].strftime("%Y-%m-%d %H:%M:%S")), end="\t")
        print(f"dtw {row[1]: 5.02f}", end="\t")
        print(f"created_by {str(row[2]).rjust(15)}", end="\t")
        print("creation_date {}".format(row[3].strftime("%Y-%m-%d %H:%M:%S")), end="\t")
        print(f"modified_by {str(row[4]).rjust(15)}", end="\t")
        if row[5]:
            print(
                "modified_date {}".format(row[5].strftime("%Y-%m-%d %H:%M:%S")),
                end="\n",
            )
        else:
            print()
```
