## Analysis results

```python
wells = db.wells_in_groups(["BAROSSA"])
dh_nos = "(" + ",".join([str(d) for d in wells.dh_no][:]) + ")"
```

```python
all_dh_nos = db.drillholes_all().dh_no
```

```python
dh_nos = "(" + ",".join([str(d) for d in all_dh_nos]) + ")"
```

```python
chem_sql = f"""
SELECT          dh.drillhole_no AS dh_no, 
                dh.unit_no      AS unit_long, 
                To_char(dh.map_100000_no) 
                                || '-' 
                                || To_char(dh.dh_seq_no) AS unit_hyphen, 
                Trim(To_char(dh.obs_well_plan_code)) 
                                || Trim(To_char(dh.obs_well_seq_no, '000')) AS obs_no, 
                dh.dh_name, 
                dh.amg_easting      AS easting, 
                dh.amg_northing     AS northing, 
                dh.amg_zone         AS zone, 
                dh.neg_lat_deg_real AS latitude, 
                dh.long_deg_real    AS longitude, 
                summ.aq_subaq       AS aquifer, 
                ar.sample_no, 
                s.collected_date, 
                s.temperature AS sample_temp, 
                ar.sample_type, 
                s.series_type, 
                s.anomalous_ind, 
                s.test_place_code  AS test_place, 
                s.extr_method_code AS extract_method, 
                s.measured_during, 
                s.collected_by, 
                s.collected_by_company AS collected_company, 
                s.data_source_code     AS data_source, 
                ar.sample_analysis_no, 
                ar.chem_code            AS chem_code, 
                ard.chem_name           AS chem_name, 
                ar.chem_value           AS value, 
                ar.chem_unit_code       AS unit, 
                meth.mnemonic           AS analysis_method_code, 
                meth.laboratory_code    AS lab, 
                meth.chem_method_desc   AS analysis_method, 
                meth.digestion_code     AS digestion, 
                meth.determination_code AS determination, 
                lims.ldl_ppm            AS lower_dl_ppm, 
                lims.udl_ppm            AS upper_dl_ppm, 
                lims.precision_ppm      AS precision_ppm, 
                ar.microanalysis_no, 
                ar.created_by, 
                ar.creation_date, 
                ar.modified_by, 
                ar.modified_date 
FROM            dhdb.sm_analysis_result_vw ar 
left outer join dhdb.sm_sample_vw s 
ON              ar.sample_no = s.sample_no 
left outer join dhdb.sm_chem_method_vw meth 
ON              ar.chem_method_no = meth.chem_method_no 
left outer join dhdb.sm_chem_vw ard 
ON              ar.chem_code = ard.chem_code 
left outer join dhdb.sm_chem_meth_chem_vw lims 
ON              ar.chem_method_no = lims.chem_method_no 
AND             ar.chem_code = lims.chem_code 
left outer join dhdb.dd_drillhole_vw dh 
ON              s.drillhole_no = dh.drillhole_no 
inner join      dhdb.dd_drillhole_summary_vw summ 
ON              dh.drillhole_no = summ.drillhole_no 
"""
```

```python
# df = db.query(chem_sql + 'WHERE dh.drillhole_no IN {dh_nos}')
df = db.query(
    chem_sql + f"WHERE ar.chem_code = 'CFC11' AND dh.drillhole_no in {dh_nos}"
)
```

```python
df.pipe(gd.cleanup_columns).dropna(how="all", axis="columns")
```
