```python
from langmetrics.llmfactory import LLMFactory
```

```python
from langmetrics.llmtestcase import LLMTestCase

simple_testcase = LLMTestCase(input='hi')
```

```python
from dotenv import load_dotenv
```

```python
load_dotenv(override=True)
```

```python
# LLM 모델 생성
openai_llm = LLMFactory.create_llm('gpt-4o-mini')
deepseek_llm = LLMFactory.create_llm('deepseek-v3')
```

```python
from langmetrics.metrics import ArenaMetric, ArenaTemplate
```

```python
# model_vs_model 모드 예시
metric = ArenaMetric(
    score_model=openai_llm,    # 채점 모델
    output_model=deepseek_llm,  # 첫 번째 응답 생성 모델
    context_model=openai_llm,  # 두 번째 응답 생성 모델
    arena_mode='model_vs_model'
)
```

```python
simple_testcase = LLMTestCase(input='만성콩팥병에 대해서 설명해줘.')
```

```python
result = await metric.ameasure(simple_testcase)
```
