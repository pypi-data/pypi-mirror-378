<dl class="py method">
                    <dt class="sig sig-object py" id="pypontem.ppl.pplparser.pplParser.extract_trend">
                      <p class="caption" role="heading"><span class="caption-text"><b>extract_trend</b></span></p>
                    <dt>
                    <dd><p>Extracts and processes trend data from an input matrix, performing unit conversions and time filtering.</p>
                    </dd></dl>

<p>1. Extracting profiles from a dataframe created using a csv file for use in a PPL file</p>

```python
import pandas as pd
input_data = pd.read_csv('data/ppl_matrix.csv')
trends = ppl.extract_trend(input_data)
print(trends.head(5))
```

<p>2. Extracting profiles from a dataframe created using a dictionary for use in a PPL file</p>

```python
matrix = {
    "varname":"PT",
    "branchname":"ISABELA_FLOWLINE",
    "out_unit":None,
    "out_unit_profile":"m",
    "time_unit":"h",
    "start_time":0,
    "end_time":2,
}

matrix_df = pd.DataFrame(matrix, index=[0])
trends = ppl.extract_trend(matrix_df)
print(trends.head(5))
```

<dl class="py method">
                    <dt class="sig sig-object py" id="pypontem.ppl.pplparser.pplParser.extract_trends_join_nodes">
                      <p class="caption" role="heading"><span class="caption-text"><b>extract_trends_join_nodes</b></span></p>
                    <dd><p>Extracts and processes trend data for branches, combining boundary and section data.</p>
                    </dd></dl>

<p>1. Joining nodes of extracted profiles from a dataframe created using a csv file for use in a PPL file</p>

```python
branch_matrix = pd.read_csv('data/branch_matrix.csv')
join_nodes = ppl.extract_trends_join_nodes(input_data,branch_matrix)
print(join_nodes.head(5))
```

<p>2. Joining nodes of extracted profiles from a dataframe created using a dictionary for use in a PPL file</p>

```python
branch_matrix = {
    "branch_in":["ISABELA_FLOWLINE","GV WELL JUMPER"],
    "branch_out":["GV WELL JUMPER","GV WELLBORE"]
}

branch_matrix_df = pd.DataFrame(branch_matrix)
joining_nodes = ppl.extract_trends_join_nodes(input_data,branch_matrix_df)
print(joining_nodes.head(2))
```
