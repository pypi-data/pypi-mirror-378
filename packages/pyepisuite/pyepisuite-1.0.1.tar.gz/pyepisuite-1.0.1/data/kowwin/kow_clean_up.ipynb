```python
import pandas as pd
```

```python
# read sheet 1 from the excel file
df_params = pd.read_excel("SrcKowData2.xlsx", sheet_name=0)
df_kow = pd.read_excel("SrcKowData2.xlsx", sheet_name=1)
df_kow_zwitter = pd.read_excel("SrcKowData2.xlsx", sheet_name=2)
```

```python
# remove the preceding 0s from each entry of the CAS Number column
df_kow["CAS Number"] = df_kow["CAS Number"].apply(lambda x: x.lstrip("0"))
df_kow_zwitter["CAS Number"] = df_kow_zwitter["CAS Number"].apply(
    lambda x: x.lstrip("0")
)
# rename the CAS Number column to CASRN
df_kow = df_kow.rename(columns={"CAS Number": "CASRN"})
df_kow_zwitter = df_kow_zwitter.rename(columns={"CAS Number": "CASRN"})
# rename Exp LogKow	and Kowwin Est to logKow_exp and logKow_pred
df_kow = df_kow.rename(
    columns={"Exp LogKow": "logKow_exp", "Kowwin Est": "logKow_pred"}
)
df_kow_zwitter = df_kow_zwitter.rename(
    columns={"Exp LogKow": "logKow_exp", "Kowwin Est": "logKow_pred"}
)
# save the dataframes to csv files
df_params.to_csv("params.csv", index=False)
df_kow.to_csv("kow.csv", index=False)
df_kow_zwitter.to_csv("kow_zwitterionic.csv", index=False)
```
