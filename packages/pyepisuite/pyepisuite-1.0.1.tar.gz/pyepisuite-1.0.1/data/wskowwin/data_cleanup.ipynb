```python
import pandas as pd
import re

def remove_leading_zeros(cas_number):
    """
    Removes leading zeros from a CAS number.

    Parameters:
    - cas_number (str): The original CAS number (e.g., '000071-41-0').

    Returns:
    - str: The CAS number without leading zeros (e.g., '71-41-0').
    """
    # Use regex to replace leading zeros before the first hyphen
    return re.sub(r'^0+(\d+-\d+-\d+)$', r'\1', cas_number)

# read the space separated cav file solubility_data.csv using pandas with separators '  +'
# note: the '  +' is a regular expression that matches two or more spaces
# we need it because the file is separated by multiple spaces, and chemical names can contain single spaces
data = pd.read_csv('solubility_data.csv', sep='  +', engine='python')
data_class = pd.read_csv('compound_classification.csv', header=None)
data_class.columns = ['class', 'name']
data_class_dict = data_class.set_index('class').to_dict()['name']
# add a new column class_name to the data DataFrame
# use the data_class_dict to map the class to the class name
data['class_name'] = data['CLASS'].map(data_class_dict)
# remove leading zeros from the CAS number
data['CAS'] = data['CAS'].apply(remove_leading_zeros)
# save data to a new csv file solubility_data_clean.csv
data.to_csv('solubility_data_clean.csv', index=False)
```
