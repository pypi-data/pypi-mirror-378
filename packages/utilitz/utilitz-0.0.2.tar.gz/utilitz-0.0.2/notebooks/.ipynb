```python
import pandas as pd
import re
from src.utilitz.excel import decode_column, encode_column
```

```python
def read_excel_table(io,
                     sheet_name=0,
                     usecols=None,
                     header=0,
                     nrows=None,
                     checkcol=None,
                     patterncol=None,
                     findtable=False,
                     raw_df=None,
                     **kwargs):
    if raw_df is None:
        raw_df = pd.read_excel(io, sheet_name=sheet_name)
    if nrows is None:
        max_nrows = float('inf')

    if header is None and findtable:
        pass

    if checkcol is not None:
        nrows = 0
        check_column = raw_df.iloc[header:, decode_column(checkcol)]

        for x in check_column:
            if not pd.isna(x) and nrows < max_nrows:
                if patterncol and re.match(patterncol, str(x)) or not patterncol:
                    nrows += 1
            else:
                break

    return pd.read_excel(io,
                         sheet_name=sheet_name,
                         usecols=usecols,
                         header=header,
                         nrows=nrows,
                         **kwargs)
```

```python
io = 'data/prueba.xlsx'
read_excel_table(io,
                 usecols='B:G',  # 'B:G', 'B:G', 'B:E'
                 header=None,  # 3, 706, 1150
                 # nrows=None,  # 20, 30, 40
                 checkcol='B',
                 # patterncol='^S'
                 )
```
