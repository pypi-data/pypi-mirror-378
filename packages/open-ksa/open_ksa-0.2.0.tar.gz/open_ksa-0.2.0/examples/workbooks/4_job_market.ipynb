```python
import pandas as pd
import os
import re
import numpy as np
# List the files in the csv subdirectory of the ministry
files_path = "opendata/ministry_of_human_resources_and_social_development/csv"
csv_files = os.listdir(files_path)

filtered_files = [file for file in csv_files if re.search(r'2025|2024', file, re.IGNORECASE)]
print(filtered_files)
print(len(filtered_files))
```

```python
hr_1 = pd.read_csv(f"{files_path}/{filtered_files[1]}")
print(filtered_files[0])
# Create an index of the columns that are arabic and the ones that are english. 1 = arabic, 0 = english 
language_index = np.array([1 if re.search(r'[ع-ي]', col) else 0 for col in hr_1.columns])

print(language_index)
hr_1
```
